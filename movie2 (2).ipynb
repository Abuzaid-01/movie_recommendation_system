```python
# Text preprocessing and feature engineering
import re
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def clean_text(text):
    """Clean and preprocess text data"""
    if pd.isna(text):
        return ""
    
    # Convert to lowercase
    text = str(text).lower()
    
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Remove extra whitespace
    text = ' '.join(text.split())
    
    return text

# Clean the overview text
file['cleaned_overview'] = file['overview'].apply(clean_text)

# Create a combined feature by combining relevant text fields
# If you have genre information, add it here
file['combined_features'] = file['cleaned_overview'].astype(str)


print(file['cleaned_overview'].iloc[0])
```

```python
# from sklearn.feature_extraction.text import CountVectorizer
# cv = CountVectorizer(max_features=6000,stop_words='english')
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(
    max_features=5000,
    stop_words='english',
    ngram_range=(1, 2),  # Include bigrams for better context
    min_df=2,  # Remove very rare words
    max_df=0.8  # Remove very common words
)
```

```python
vector = cv.fit_transform(file["overview"]).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie, num_recommendations=5):
    # Case-insensitive search
    movie_matches = file[file['Title'].str.lower() == movie.lower()]
    
    if movie_matches.empty:
        print(f"Movie '{movie}' not found in the database")
        # Show similar titles
        similar = file[file['Title'].str.contains(movie, case=False, na=False)]['Title'].head(5)
        if not similar.empty:
            print("Did you mean one of these?")
            for title in similar:
                print(f"- {title}")
        return
    
    index = movie_matches.index[0]
    movie_title = movie_matches.iloc[0]['Title']
    
    print(f"Movies similar to '{movie_title}':")
    print("-" * 50)
    
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    
    for i, (idx, score) in enumerate(distances[1:num_recommendations+1], 1):
        title = file.iloc[idx].Title
        rating = file.iloc[idx].vote_average
        print(f"{i}. {title} (Rating: {rating:.1f}, Similarity: {score:.3f})")
```

```python
def search_movies(query, limit=10):
    """Search for movies by partial title match"""
    matches = file[file['Title'].str.contains(query, case=False, na=False)]
    
    if matches.empty:
        print(f"No movies found containing '{query}'")
        return
    
    print(f"Found {len(matches)} movies containing '{query}':")
    print("-" * 50)
    
    for idx, (_, movie) in enumerate(matches.head(limit).iterrows(), 1):
        print(f"{idx}. {movie['Title']} (Rating: {movie['vote_average']:.1f})")

# Test it
search_movies('batman')
```

```python
import pickle
import os

# Create a models directory
os.makedirs('models', exist_ok=True)

# Save the processed dataframe
file.to_pickle('models/movie_data.pkl')

# Save the CountVectorizer (fitted model)
with open('models/count_vectorizer.pkl', 'wb') as f:
    pickle.dump(cv, f)

# Save the similarity matrix
with open('models/similarity_matrix.pkl', 'wb') as f:
    pickle.dump(similarity, f)

# Save just the essential data for faster loading
movie_info = file[['Title', 'overview', 'vote_average', 'poster_path']].copy()
movie_info.to_pickle('models/movie_info.pkl')

print("Models saved successfully!")
print("Saved files:")
print("- models/movie_data.pkl (full dataset)")
print("- models/count_vectorizer.pkl (fitted vectorizer)")
print("- models/similarity_matrix.pkl (similarity matrix)")
print("- models/movie_info.pkl (essential movie info)")
```
